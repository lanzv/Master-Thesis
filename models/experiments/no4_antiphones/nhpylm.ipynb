```python
from src.utils.loader import prepare_dataset
from src.eval.pipelines import evaluation_pipeline
from nhpylm_model import NHPYLMModel
# no4
# Get Data
X, y = prepare_dataset()
X_train, y_train = X[:int(0.9*9392)].tolist(), y[:int(0.9*9392)].tolist()
X_dev, y_dev = X[int(0.9*9392):9392].tolist(), y[int(0.9*9392):9392].tolist()
X_test, y_test = X[9392:].tolist(), y[9392:].tolist()
# Init model
model = NHPYLMModel(7)
# Train and Fit model
model.train(X_train, X_dev, y_train, y_dev, 50, True, True)
# Predictions
mjwp_score = model.get_mjwp_score()
X_train = X_train+X_dev
y_train = y_train+y_dev
train_segmentation, train_perplexity = model.predict_segments(X_train)
test_segmentation, test_perplexity = model.predict_segments(X_test)

# Evaluate model
bacor_model = evaluation_pipeline(
    train_segmentation, y_train, test_segmentation, y_test, train_perplexity, test_perplexity, mjwp_score,
    max_features_from_model = 100, include_additative = False, fe_occurence_coef=5)
```
